```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Plotting vectors

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,-6,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

## Addition of two vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

## Subtraction of two vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
diff = vector_1 - vector_2
print(diff)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

## Multiply a vector by a scaler

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = 2 * vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,4,6,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = -0.5 * vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0.,0.,-1.,-1.5,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

##Dot product of two vectors

```python
a = np.array([2,3])
b = np.array([4,4])
a_dot_b = np.dot(a,b)
print(a_dot_b)
```

```python
c = np.array([40,20,35])
d = np.array([53,24,68])
c_dot_d = np.dot(c,d)
print(c_dot_d)
```

## cross product of two vectors

```python
a = np.array([2,3])
b = np.array([4,4])
a_cross_b = np.cross(a,b)
print(a_cross_b)
```

```python
a = np.array([2,3,6])
b = np.array([4,4,7])
a_cross_b = np.cross(a,b)
print(a_cross_b)
```

## projection of vectors

```python
a = np.array([2,5])
v = np.asarray([8,-6])

# magnitude of "V" vector
magnitude =np.sqrt(np.sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/magnitude**2)*v

print("Projection of a vector on v vector ",proj_of_a_on_v)
```
