## Matrix subtraction

```python
A = np.array([[1,2],[3,4]])
B = np.array([[3,2],[7,9]])
difference = A - B
print(difference)
```

```python
matrix1 = np.random.randint(10,size = (3,3))
matrix2 = np.random.randint(10,size = (3,3))
print("Matrix 1 : ",matrix1)
print("Matrix 2 : ",matrix2)
diff = matrix1 - matrix2
print("Difference  :", diff)
```

## Multipy matrix by a scalar value

## Multiplication ot two matrices

```python
A = np.array([[1,2],[3,4]])
B = np.array([[3,2],[7,9]])
ans = np.dot(A,B)
print("Answer  : ", ans)
```

```python
matrix1 = np.random.randint(10,size = (3,3))
matrix2 = np.random.randint(10,size = (3,3))
print("Matrix 1 : ",matrix1)
print("Matrix 2 : ",matrix2)
prod = np.dot(matrix1,matrix2)
print("product :",prod)
```

```python
matrix1 = np.random.randint(10,size = (3,3))
matrix2 = np.random.randint(10,size = (4,3))
print("Matrix 1 : ",matrix1)
print("Matrix 2 : ",matrix2)
prod = np.dot(matrix1,matrix2)
print("product :",prod)
```

## Element multiplication
